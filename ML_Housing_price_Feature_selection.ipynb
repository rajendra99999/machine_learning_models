# EDA housing price feature selection

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.feature_selection import SelectFromModel
from sklearn.linear_model import Lasso
pd.pandas.set_option('display.max_columns', None)
```

```python
dataset=pd.read_csv('train_HP_clnd.csv')
```

```python
dataset.head()
```

```python
from sklearn.model_selection import train_test_split

y=dataset['SalePrice']

x=dataset.drop(['SalePrice','Id'],axis=1)

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.7,random_state=0)
```

```python
lasso=Lasso(random_state=0,alpha=0.05)
```

```python
feature_selection=SelectFromModel(lasso)
```

```python
feature_selection.fit(x_train,y_train)
```

```python
feature_selection.get_support()
```

```python
seleceted_feature=x_train.columns[(feature_selection.get_support())]
```

```python
seleceted_feature
```
